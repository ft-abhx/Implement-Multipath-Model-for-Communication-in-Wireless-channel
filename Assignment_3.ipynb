# 1. MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
X=np.array([[1,2,3,4,5,6,500],[-100,-50,0,50,100,2000,10000],[2,1,1,1,1,1,1]]) #3 Samples, 7 attributes or features
print(X)
print(X.shape)
scaler = MinMaxScaler()
scaler.fit(X)
xscale=scaler.transform(X)

print(xscale)
```

**Min Max Scaling**
$$xscale=\frac{x-x_{min}}{x_{max}-x_{min}}$$
**Example**

How 0.99019608 is obtained?

Values in first columns are $1,-100,2$. $x_{min}=-100$,$x_{max}=2$

Calculating xscale for first value, that is $x=1$
$$xscale=\frac{1-(-100)}{2-(-100)}$$
which means
$$xscale=\frac{101}{102}$$

```python
101/102
```

# 2. RobustScaler

```python
#Write you code here illustrating how values are transformed using RobustScaler

from sklearn.preprocessing import RobustScaler

#3 Samples, 7 attributes or features
X=np.array([[1,2,3,4,5,6,500],[-100,-50,0,50,100,2000,10000],[2,1,1,1,1,1,1]]) 
print("Array: ",X)
print("\n Dimensions: ",X.shape)

scaler = RobustScaler()
scaler.fit(X)#Compute the minimum and maximum to be used for later scaling
xscale=scaler.transform(X)
print("\n", xscale)
```

```python
import numpy as np

x1 = np.array([1, -100, 2]) # your 1st column here

q75, q25 = np.percentile(x1, [75 ,25])
iqr = q75 - q25 #interquartile range

print("x75: ",q75)
print("x25: ",q25)

x1_med = np.median(x1)
print("xmed: ",x1_med)

x1_scaled = (x1-x1_med)/iqr
x1_scaled
```

Explain the process of Robust Scaling with an example

**Robust Scaler**
$$xscale=\frac{xi-x_{med}}{x_{75}-x_{25}}$$

***For example:***

Values in first columns are $1,-100,2$. $x_{75}=1.5$, $x_{25}=-49.5$, $x_{med}=1$

Calculating xscale for first value, that is $x=1$
$$xscale=\frac{1-(1)}{1.5-(-49.5)}$$
which means
$$xscale=\frac{0}{51}$$

# 3. StandardScaler

```python
#Write your code here illustrating how values are transformed using StandardScale

from sklearn.preprocessing import StandardScaler
import pandas as pd

X=np.array([[6, 2], [4, 2], [6, 4], [8, 2]])

print("Array: ",X)
print("\n Dimensions: ",X.shape)

scaler = StandardScaler()
scaler.fit(X) 
xscale=scaler.transform(X)
print("\n",xscale)
```

```python
std=np.std(X, axis=0)
mean=np.mean(X,axis=0)
print("std: ",std)
print("mean: ",mean)
```

Explain the process of Standard Scaling with an example

**Standard Scaler**
$$xscale=\frac{xi-x_{mean}}{x_{StandardDeviation}}$$

***For Example:***


At (0,1):

$x=2$, $x_{mean}=2.5$, $x_{StandardDeviation}=0.8660254$

Calculating xscale , that is $x=2$
$$xscale=\frac{2-(2.5)}{0.8660254}$$
which means
$$xscale=\frac{-0.5}{0.8660254}$$

# 4. Normalizer

```python
#Write your code here illustrating how values are transformed using Normalizer

from sklearn.preprocessing import Normalizer
X=np.array([[6, 2], [4, 2], [6, 4], [8, 2]])

print("Array: ",X)
print("\n Dimensions: ",X.shape)

scaler = Normalizer()
scaler.fit(X) 
xscale=scaler.transform(X)
print("\n",xscale)
```

Explain the process of Normalizer with an example

**Normalizer**

Taking the row [6,2],

First, we need to take the sqareroot of the sum of the sqaures of the element in the row.
$sq = \sqrt{6^2 + 2^2}$

To get the normalised row, we need to divide each element of the row with sq

$xscale = 6/\sqrt{6^2 + 2^2}$

```python
import math
sq=math.sqrt(6*6+2*2)
xscale=6/sq

print("Sq: ",sq)
print("xscale: ",xscale)
```
